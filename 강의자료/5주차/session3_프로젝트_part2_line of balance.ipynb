```python
import numpy as np
from itertools import *
import collections
import pandas as pd
```

```python
# 데이터 가져오기
df = pd.read_excel('project_lob.xlsx')
df
```

```python
# asis만 따로 dataframe 만들기
df_asis = df[['작업','As-is']][:9]
df_asis
```

```python
# asis 데이터로 필요한 계산식 작성
실동시간_1일 = 360
작업목표량 = 3000
소요시간합계 = df["As-is"].sum()
애로작업시간 = df["As-is"].max()
작업수 = df["As-is"].count()
라인공수 = 작업수*애로작업시간
라인편성효율 = 100*(소요시간합계/라인공수)
일작업능력 = 실동시간_1일/애로작업시간
인당_일작업량 = 일작업능력/작업수
TT = 실동시간_1일/작업목표량
인원수 = 소요시간합계/TT
연간인건비 = 30*인원수
data = {
    '변수': ['구분','소요시간합계', '애로작업시간', '작업수', '라인공수', '라인편성효율', '일작업능력', '인당_일작업량', '작업목표량', 'TT', '인원수', '연간인건비', '절감액', '투자금액','회수기간' ],
    '합계': ['As-is',소요시간합계, 애로작업시간, 작업수, 라인공수, 라인편성효율, 일작업능력, 인당_일작업량,작업목표량, TT, 인원수, 연간인건비,0, 0, 0]
}
df_asis = pd.DataFrame(data)
df_asis
```

```python
# asis 연간인건비 변수 지정
연간인건비_asis = df_asis.iloc[11, 1]
연간인건비_asis
```

```python
# tobe 계산식 for loop 구문 작성
result = []
for i in range(2,7):
  실동시간_1일 = 360
  작업목표량 = 3000
  소요시간합계 = df.iloc[0:9,i].sum()
  애로작업시간 = df.iloc[0:9,i].max()
  작업수 = df.iloc[0:9,i].count()
  라인공수 = 작업수*애로작업시간
  라인편성효율 = 100*(소요시간합계/라인공수)
  일작업능력 = 실동시간_1일/애로작업시간
  인당_일작업량 = 일작업능력/작업수
  TT = 실동시간_1일/작업목표량
  인원수 = 소요시간합계/TT
  연간인건비 = 30*인원수
  절감액 = 연간인건비_asis-연간인건비
  투자금액 = df.iloc[9,i]
  회수기간 = 투자금액/절감액
  data = [['tobe' + str(i-1),소요시간합계, 애로작업시간, 작업수, 라인공수, 라인편성효율, 일작업능력, 인당_일작업량,작업목표량, TT, 인원수, 연간인건비, 절감액, 투자금액, 회수기간]]

  df_tobe = pd.DataFrame(data)
  result.append(df_tobe)

df_tobe = pd.concat(result, axis=0)
df_tobe = df_tobe.transpose()
df_tobe
```

```python
# asis와 tobe dataframe 결합
final_df = pd.concat([df_asis, df_tobe], axis=1)
final_df = final_df.transpose()
final_df
```

```python
# dataframe 정리
new_header = final_df.iloc[0]
final_df = final_df[1:]
final_df.columns = new_header
final_df.set_index('구분')
```
